# 1. Segmentação por Experiência com Tecnologia

```python
# 1. Segmentação por Experiência com Tecnologia
def segmentar_tecnologia(row):
    if row['Ferramentas_Anteriores'] == 'Sim' and row['Experiencia_LLMs'] == 'Sim' and row['Utilidade_LLMs'] in [5, 4]:
        return 'Usuários Entusiasmados de Tecnologia'
    elif row['Ferramentas_Anteriores'] == 'Não' and row['Experiencia_LLMs'] == 'Não' and row['Utilidade_LLMs'] in [1, 2]:
        return 'Céticos da Tecnologia'
    else:
        return 'Usuários Ocasionais'

df_questionario['Segmento_Tecnologia'] = df_questionario.apply(segmentar_tecnologia, axis=1)

# Exibir as primeiras linhas do DataFrame com as novas colunas de segmento
print(df_questionario.head())
```

```python
# Variáveis categóricas
variaveis_categoricas = ['Pos_Graduacao', 'Instituicao', 'Frequencia_Elaboracao', 'Ferramentas_Anteriores', 'Experiencia_LLMs', 'Utilidade_LLMs']
```

```python
# 1. Comparar as Pontuações Médias dos Critérios de Avaliação entre os Segmentos
colunas_criterios = [f'{i} {c}' for i in range(1, 21)
                          for c in ['Cobertura', 'Gramaticalidade', 'Utilidade', 'Relevância', 'Capacidade de resposta']]

medias_por_segmento = df_questionario.groupby('Segmento_Tecnologia')[colunas_criterios].mean()
print("\nMédias das pontuações por segmento:\n", medias_por_segmento)

# Criar um gráfico de barras comparativo para um critério específico
criterio_exemplo = '1 Cobertura'
plt.figure(figsize=(10, 6))
medias_por_segmento[criterio_exemplo].plot(kind='bar')
plt.title(f'Média de {criterio_exemplo} por Segmento de Tecnologia')
plt.xlabel('Segmento de Tecnologia')
plt.ylabel('Pontuação Média')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2. Testes de Significância Estatística (ANOVA)
# Exemplo para um critério específico
grupo1 = df_questionario[df_questionario['Segmento_Tecnologia'] == 'Usuários Entusiasmados de Tecnologia'][criterio_exemplo]
grupo2 = df_questionario[df_questionario['Segmento_Tecnologia'] == 'Céticos da Tecnologia'][criterio_exemplo]
grupo3 = df_questionario[df_questionario['Segmento_Tecnologia'] == 'Usuários Ocasionais'][criterio_exemplo]

# Realizar o teste ANOVA
fvalue, pvalue = stats.f_oneway(grupo1, grupo2, grupo3)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue:.2f}')
print(f'  Valor p: {pvalue:.3f}')

# Interpretar o resultado
if pvalue < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```

```python
# 3. Analisar a Distribuição das Pontuações por Segmento
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segmento_Tecnologia', y=criterio_exemplo, data=df_questionario)
plt.title(f'Boxplot de {criterio_exemplo} por Segmento de Tecnologia')
plt.xlabel('Segmento de Tecnologia')
plt.ylabel('Pontuação')
plt.show()
```

```python
# 4. Analisando as variaveis categoricas nos seguimentos
for var in variaveis_categoricas:
    print(f'Analisando a coluna: {var}')
    if df_questionario[var].dtype == 'object':  # Variável categórica
        print(df_questionario[var].value_counts())
        plt.figure(figsize=(10, 6))
        sns.countplot(x=df_questionario[var], hue='Segmento_Tecnologia', data=df_questionario, order = df_questionario[var].value_counts().index)
        plt.title(f'Distribuição de {var}')
        plt.xticks(rotation=45, ha='right')
        plt.show()
```

# 2. Segmentação por Nível de Estresse e Satisfação Profissional:

```python
# 2. Segmentação por Nível de Estresse e Satisfação Profissional
def segmentar_estresse_satisfacao(row):
    # Critérios para segmentação:
    # "Satisfeitos e Relaxados": Baixo estresse (1-2) e alta satisfação (4-5)
    # "Estressados e Insatisfeitos": Alto estresse (4-5) e baixa satisfação (1-2)
    # "Equilibrados": Níveis médios de estresse (3) e satisfação (3)

    if (row['Nivel_Estresse'] in [1, 2]) and (row['Satisfacao_Profissional'] in [4, 5]):
        return 'Satisfeitos e Relaxados'
    elif (row['Nivel_Estresse'] in [4, 5]) and (row['Satisfacao_Profissional'] in [1, 2]):
        return 'Estressados e Insatisfeitos'
    else:
        return 'Equilibrados'

# Aplicar a segmentação
df_questionario['Segmento_Estresse_Satisfacao'] = df_questionario.apply(segmentar_estresse_satisfacao, axis=1)
```

```python
# Visualizar a distribuição dos segmentos
plt.figure(figsize=(10, 6))
df_questionario['Segmento_Estresse_Satisfacao'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Segmentos de Estresse e Satisfação')
plt.xlabel('Segmento')
plt.ylabel('Número de Professores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Análise comparativa das pontuações médias dos critérios por segmento
medias_por_segmento_estresse = df_questionario.groupby('Segmento_Estresse_Satisfacao')[colunas_criterios].mean()
print("\nMédias das pontuações por segmento de Estresse e Satisfação:\n", medias_por_segmento_estresse)
```

```python
# Exemplo de gráfico comparativo para um critério específico
criterio_exemplo = '1 Cobertura'
plt.figure(figsize=(10, 6))
medias_por_segmento_estresse[criterio_exemplo].plot(kind='bar')
plt.title(f'Média de {criterio_exemplo} por Segmento de Estresse e Satisfação')
plt.xlabel('Segmento de Estresse e Satisfação')
plt.ylabel('Pontuação Média')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Teste de significância estatística (ANOVA)
grupo1 = df_questionario[df_questionario['Segmento_Estresse_Satisfacao'] == 'Satisfeitos e Relaxados'][criterio_exemplo]
grupo2 = df_questionario[df_questionario['Segmento_Estresse_Satisfacao'] == 'Estressados e Insatisfeitos'][criterio_exemplo]
grupo3 = df_questionario[df_questionario['Segmento_Estresse_Satisfacao'] == 'Equilibrados'][criterio_exemplo]

# Realizar o teste ANOVA
fvalue, pvalue = stats.f_oneway(grupo1, grupo2, grupo3)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue:.2f}')
print(f'  Valor p: {pvalue:.3f}')

# Interpretar o resultado
if pvalue < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```

```python
# Boxplot para visualizar a distribuição das pontuações
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segmento_Estresse_Satisfacao', y=criterio_exemplo, data=df_questionario)
plt.title(f'Boxplot de {criterio_exemplo} por Segmento de Estresse e Satisfação')
plt.xlabel('Segmento de Estresse e Satisfação')
plt.ylabel('Pontuação')
plt.show()
```

```python
# Correlação entre Nível de Estresse, Satisfação Profissional e pontuações dos critérios
# Calcular a correlação média para todos os critérios
correlacoes = {}
for criterio in colunas_criterios:
    correlacoes[criterio] = {
        'Correlacao_Estresse': df_questionario['Nivel_Estresse'].corr(df_questionario[criterio]),
        'Correlacao_Satisfacao': df_questionario['Satisfacao_Profissional'].corr(df_questionario[criterio])
    }
```

```python
# Criar um DataFrame para visualizar as correlações
df_correlacoes = pd.DataFrame.from_dict(correlacoes, orient='index')
print("\nCorrelações entre Nível de Estresse/Satisfação e Critérios:")
print(df_correlacoes)
```

```python
# Mapa de calor das correlações
plt.figure(figsize=(12, 8))
sns.heatmap(df_correlacoes, annot=True, cmap='coolwarm', center=0)
plt.title('Correlações entre Estresse/Satisfação e Critérios de Avaliação')
plt.tight_layout()
plt.show()
```

#3. Segmentação por Tempo de Experiência:

```python
# 3. Segmentação por Tempo de Experiência:
def segmentar_experiencia(row):
    """
    Segmentar os professores em:
    - "Novatos": menos de 5 anos de experiência
    - "Experientes": de 5 a 20 anos de experiência
    - "Veteranos": mais de 20 anos de experiência
    """
    try:
        tempo = float(row['Tempo_Experiencia_Numerico'])
    except:
        # Caso não seja possível converter, assume 0 anos de experiência
        tempo = 0.0

    if tempo < 5:
        return 'Novatos'
    elif tempo <= 20:
        return 'Experientes'
    else:
        return 'Veteranos'

# Aplicar a segmentação
df_questionario['Segmento_Experiencia'] = df_questionario.apply(segmentar_experiencia, axis=1)

# Visualizar a distribuição dos segmentos de tempo de experiência
plt.figure(figsize=(10, 6))
df_questionario['Segmento_Experiencia'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Segmentos por Tempo de Experiência')
plt.xlabel('Segmento de Experiência')
plt.ylabel('Número de Professores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Comparar as pontuações médias dos critérios por segmento de experiência
medias_por_segmento_exp = df_questionario.groupby('Segmento_Experiencia')[colunas_criterios].mean()
print("\nMédias das pontuações por segmento de Tempo de Experiência:\n", medias_por_segmento_exp)
```

```python
# Exemplo de gráfico comparativo para um critério específico (utilizando "1 Cobertura")
criterio_exemplo = '1 Cobertura'
plt.figure(figsize=(10, 6))
medias_por_segmento_exp[criterio_exemplo].plot(kind='bar')
plt.title(f'Média de {criterio_exemplo} por Segmento de Tempo de Experiência')
plt.xlabel('Segmento de Experiência')
plt.ylabel('Pontuação Média')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Teste de significância estatística (ANOVA) para o critério escolhido
grupo_novatos = df_questionario[df_questionario['Segmento_Experiencia'] == 'Novatos'][criterio_exemplo]
grupo_experientes = df_questionario[df_questionario['Segmento_Experiencia'] == 'Experientes'][criterio_exemplo]
grupo_veteranos = df_questionario[df_questionario['Segmento_Experiencia'] == 'Veteranos'][criterio_exemplo]

fvalue_exp, pvalue_exp = stats.f_oneway(grupo_novatos, grupo_experientes, grupo_veteranos)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue_exp:.2f}')
print(f'  Valor p: {pvalue_exp:.3f}')

if pvalue_exp < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```

```python
# Boxplot para visualizar a distribuição das pontuações nos segmentos de experiência
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segmento_Experiencia', y=criterio_exemplo, data=df_questionario)
plt.title(f'Boxplot de {criterio_exemplo} por Segmento de Tempo de Experiência')
plt.xlabel('Segmento de Tempo de Experiência')
plt.ylabel('Pontuação')
plt.show()
```

```python
# (Opcional) Análise de correlação entre o Tempo de Experiência e os critérios de avaliação
correlacoes_exp = {}
for criterio in colunas_criterios:
    correlacoes_exp[criterio] = df_questionario['Tempo_Experiencia_Numerico'].corr(df_questionario[criterio])

df_correlacoes_exp = pd.DataFrame.from_dict(correlacoes_exp, orient='index', columns=['Correlacao_Tempo_Experiencia'])
print("\nCorrelação entre Tempo de Experiência e Critérios:")
print(df_correlacoes_exp)

plt.figure(figsize=(12, 8))
sns.heatmap(df_correlacoes_exp, annot=True, cmap='coolwarm', center=0)
plt.title('Correlação entre Tempo de Experiência e Critérios de Avaliação')
plt.tight_layout()
plt.show()
```

# 4. Segmentação por Nível de Pós-Graduação:



```python
# 4. Segmentação por Nível de Pós-Graduação:

def segmentar_pos_graduacao(row):
    """
    Segmenta os professores por nível de pós-graduação.
    """
    pos_graduacao = row['Pos_Graduacao']
    if pos_graduacao == 'Especialização':
        return 'Especialistas'
    elif pos_graduacao == 'Mestrado':
        return 'Mestres'
    elif pos_graduacao == 'Doutorado':
        return 'Doutores'
    else:
        return 'Outro/Não Possui'  # Para lidar com outras respostas

# Aplicar a segmentação
df_questionario['Segmento_Pos_Graduacao'] = df_questionario.apply(segmentar_pos_graduacao, axis=1)

# Visualizar a distribuição dos segmentos
plt.figure(figsize=(10, 6))
df_questionario['Segmento_Pos_Graduacao'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Segmentos por Nível de Pós-Graduação')
plt.xlabel('Segmento de Pós-Graduação')
plt.ylabel('Número de Professores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Análise comparativa das pontuações médias dos critérios por segmento
medias_por_segmento_pos = df_questionario.groupby('Segmento_Pos_Graduacao')[colunas_criterios].mean()
print("\nMédias das pontuações por segmento de Pós-Graduação:\n", medias_por_segmento_pos)
```

```python
# Exemplo de gráfico comparativo para um critério específico (utilizando "1 Cobertura")
criterio_exemplo = '1 Cobertura'
plt.figure(figsize=(10, 6))
medias_por_segmento_pos[criterio_exemplo].plot(kind='bar')
plt.title(f'Média de {criterio_exemplo} por Segmento de Pós-Graduação')
plt.xlabel('Segmento de Pós-Graduação')
plt.ylabel('Pontuação Média')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Teste de significância estatística (ANOVA)
# Remover o segmento "Outro/Não Possui" para o teste ANOVA, pois pode ter poucos elementos
df_filtered = df_questionario[df_questionario['Segmento_Pos_Graduacao'].isin(['Especialistas', 'Mestres', 'Doutores'])]

grupo_especialistas = df_filtered[df_filtered['Segmento_Pos_Graduacao'] == 'Especialistas'][criterio_exemplo]
grupo_mestres = df_filtered[df_filtered['Segmento_Pos_Graduacao'] == 'Mestres'][criterio_exemplo]
grupo_doutores = df_filtered[df_filtered['Segmento_Pos_Graduacao'] == 'Doutores'][criterio_exemplo]

# Realizar o teste ANOVA
fvalue_pos, pvalue_pos = stats.f_oneway(grupo_especialistas, grupo_mestres, grupo_doutores)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue_pos:.2f}')
print(f'  Valor p: {pvalue_pos:.3f}')

# Interpretar o resultado
if pvalue_pos < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```

#5. Segmentação por Frequência de Elaboração de Questões:



```python
# 5. Segmentação por Frequência de Elaboração de Questões:

def segmentar_frequencia_elaboracao(row):
    """
    Segmenta os professores pela frequência com que elaboram questões.
    """
    frequencia = row['Frequencia_Elaboracao']
    if frequencia in ['Sempre', 'Frequentemente']:
        return 'Criadores Frequentes'
    elif frequencia in ['Às vezes', 'Raramente']:
        return 'Criadores Ocasionais'
    else:
        return 'Não Criadores'  # Para "Nunca" ou respostas inesperadas

# Aplicar a segmentação
df_questionario['Segmento_Frequencia_Elaboracao'] = df_questionario.apply(segmentar_frequencia_elaboracao, axis=1)

# Visualizar a distribuição dos segmentos
plt.figure(figsize=(10, 6))
df_questionario['Segmento_Frequencia_Elaboracao'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Segmentos por Frequência de Elaboração de Questões')
plt.xlabel('Segmento de Frequência de Elaboração')
plt.ylabel('Número de Professores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Análise comparativa das pontuações médias dos critérios por segmento
medias_por_segmento_freq = df_questionario.groupby('Segmento_Frequencia_Elaboracao')[colunas_criterios].mean()
print("\nMédias das pontuações por segmento de Frequência de Elaboração:\n", medias_por_segmento_freq)
```

```python
# Exemplo de gráfico comparativo para um critério específico (utilizando "1 Cobertura")
criterio_exemplo = '1 Cobertura'
plt.figure(figsize=(10, 6))
medias_por_segmento_freq[criterio_exemplo].plot(kind='bar')
plt.title(f'Média de {criterio_exemplo} por Segmento de Frequência de Elaboração')
plt.xlabel('Segmento de Frequência de Elaboração')
plt.ylabel('Pontuação Média')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Teste de significância estatística (ANOVA)
grupo_frequentes = df_questionario[df_questionario['Segmento_Frequencia_Elaboracao'] == 'Criadores Frequentes'][criterio_exemplo]
grupo_ocasionais = df_questionario[df_questionario['Segmento_Frequencia_Elaboracao'] == 'Criadores Ocasionais'][criterio_exemplo]
grupo_nao_criadores = df_questionario[df_questionario['Segmento_Frequencia_Elaboracao'] == 'Não Criadores'][criterio_exemplo]

# Realizar o teste ANOVA
fvalue_freq, pvalue_freq = stats.f_oneway(grupo_frequentes, grupo_ocasionais, grupo_nao_criadores)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue_freq:.2f}')
print(f'  Valor p: {pvalue_freq:.3f}')

# Interpretar o resultado
if pvalue_freq < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```

#6. Segmentação por Tipo de Instituição:


```python
# 6. Segmentação por Tipo de Instituição:
def segmentar_instituicao(row):
    """
    Segmenta os professores por tipo de instituição.
    """
    instituicao = row['Instituicao']
    if 'Municipal' in instituicao:
        return 'Público Municipal'
    elif 'Estadual' in instituicao:
        return 'Público Estadual'
    elif 'Federal' in instituicao:
        return 'Público Federal'
    elif 'Particular' in instituicao:
        return 'Particular'
    else:
        return 'Outro/Não Especificado'

# Aplicar a segmentação
df_questionario['Segmento_Instituicao'] = df_questionario.apply(segmentar_instituicao, axis=1)

# Visualizar a distribuição dos segmentos
plt.figure(figsize=(10, 6))
df_questionario['Segmento_Instituicao'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Segmentos por Tipo de Instituição')
plt.xlabel('Segmento de Instituição')
plt.ylabel('Número de Professores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Teste de significância estatística (ANOVA)
# Filtrar para remover segmentos com poucos dados, se necessário
df_filtered_inst = df_questionario[df_questionario['Segmento_Instituicao'].isin(['Público Municipal', 'Público Estadual', 'Público Federal', 'Particular'])]

grupo_municipal = df_filtered_inst[df_filtered_inst['Segmento_Instituicao'] == 'Público Municipal'][criterio_exemplo]
grupo_estadual = df_filtered_inst[df_filtered_inst['Segmento_Instituicao'] == 'Público Estadual'][criterio_exemplo]
grupo_federal = df_filtered_inst[df_filtered_inst['Segmento_Instituicao'] == 'Público Federal'][criterio_exemplo]
grupo_particular = df_filtered_inst[df_filtered_inst['Segmento_Instituicao'] == 'Particular'][criterio_exemplo]

# Realizar o teste ANOVA
fvalue_inst, pvalue_inst = stats.f_oneway(grupo_municipal, grupo_estadual, grupo_federal, grupo_particular)
print(f'\nTeste ANOVA para {criterio_exemplo}:')
print(f'  Valor F: {fvalue_inst:.2f}')
print(f'  Valor p: {pvalue_inst:.3f}')

# Interpretar o resultado
if pvalue_inst < 0.05:
    print('  Há diferenças estatisticamente significativas entre os grupos.')
else:
    print('  Não há diferenças estatisticamente significativas entre os grupos.')
```
